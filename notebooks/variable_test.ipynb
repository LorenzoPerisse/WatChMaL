```python
import numpy as np
import matplotlib.pyplot as plt
import sys
import uproot
```

## Dropout

```python
dropout = [0.2,0.1,0.05,0.01,0]
res_position = [877.41 , 526.02 , 414.75 , 356.14 , 359.20]
# res_direction = [301.89 , 301.98 , 302.16 , 302.17 , 302.22]
# res_orthodirection = [336.67 , 343.44 , 354.68 , 351.65 , 326.17]

plt.plot(dropout, res_position, marker='o', c='b', linestyle='-', label='norm')
# plt.plot(dropout, res_direction, marker='o', c='g', linestyle='-', label='direction projection')
# plt.plot(dropout, res_orthodirection, marker='o', c='r', linestyle='-', label='ortho to the direction projection')
plt.xlabel('Dropout')
plt.legend()
plt.ylabel('Resolution at 68% (in cm)')
plt.title('Resolution at 68% for position resolution in function of dropout')
```
